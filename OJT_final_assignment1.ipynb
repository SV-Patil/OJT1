### Q1. Data visualization and statistical measures.
for this question you are required to analyse the iris dataset using python. Perform all data visualization techniques on all numerical columns of the dataset. addtionally calculate all possible statistical measures for each numerical columns.

```python
import pandas as pd
import numpy as np
import scipy.stats as stat
import matplotlib.pyplot as plt
import statistics
```

```python
dataset = pd.read_csv("Data/iris.csv")
dataset
```

```python
PL = dataset["Petal.Length"]
SL = dataset["Sepal.Length"]
PW = dataset["Petal.Width"]
SW = dataset["Sepal.Width"]
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.hist(SL)
plt.xlabel("Sepal Length")

plt.subplot(2,2,2)
plt.hist(SW)
plt.xlabel("Sepal Width")

plt.subplot(2,2,3)
plt.hist(PL)
plt.xlabel("Petal Length")

plt.subplot(2,2,4)
plt.hist(PW)
plt.xlabel("Petal Width")

plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.scatter(PL,SL)

plt.subplot(2,2,2)
plt.scatter(PL,SW)

plt.subplot(2,2,3)
plt.scatter(PW,SL)

plt.subplot(2,2,4)
plt.scatter(PW,SW)
plt.show()
```

```python
lis = [PL,SL,PW,SW]
name = ["Petal Length", "Sepal Length", "Petal Width", "Sepal Width"]
for i in range(4):
    print(f"The mean of {name[i]} is: ",np.round(np.mean(lis[i]),3))
    print(f"The standard deviation of {name[i]} is: ",np.round(np.std(lis[i]),3))
    print(f"The median of {name[i]} is: ",np.round(statistics.median(lis[i]),3))
    print(f"The mode of {name[i]} is: ",np.round(statistics.mode(lis[i]),3))
    print("\n")
```
